## Python Setup

```python
import pandas as pd
```

## Read In

```python
emails = pd.read_excel('../output/email_linkage_GC.xlsx')
del emails['name_id_grnt']
profiles = pd.read_csv('../output/researcher_profiles.csv')
```

## Linkage

```python
profiles['name'] = (profiles['prenom_port'] + ' ' + profiles['nom_port']).str.upper()
del profiles['prenom_port'], profiles['nom_port']
profiles['name'] = profiles['name'].str.replace(r'\s?-\s?', ' ')
profiles['name'] = profiles['name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
linkage = pd.merge(emails, profiles, how = 'left', on = 'name')
```

```python
linkage.describe(include='all')
```

```python
linkage = linkage[(linkage['Source']=='GC')|(linkage['Source']=='Linkage')|(linkage['Source'].isnull())]
```

```python
linkage.head()
```

```python
del linkage['Source']
```

```python
linkage[linkage['emails'].notnull()].describe(include = 'all')
```

```python
linkage.to_csv('../output/researcher_info.csv', index=False)
```
